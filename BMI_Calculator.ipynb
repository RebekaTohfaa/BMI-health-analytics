# 📊 BMI Health Analytics Project

Built a BMI calculator using Python and applied basic analytical logic to classify health categories and generate insights.

## 📊 Insight

BMI helps categorize individuals into different health groups such as underweight, normal, overweight, and obese.

This simple classification demonstrates how data can be used to support basic health analysis and decision-making.

```python
import pandas as pd

data = pd.DataFrame({
    "weight": [120, 150, 180, 200, 220],
    "height": [60, 65, 70, 72, 68]
})

data["BMI"] = (data["weight"] * 703) / (data["height"] ** 2)
data

def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 24.9:
        return "Normal"
    elif bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

data["Category"] = data["BMI"].apply(bmi_category)
data
```

```python
# Average BMI
data["BMI"].mean()

# Category distribution
data["Category"].value_counts()
```

```python
import matplotlib.pyplot as plt

data["BMI"].plot(kind="hist")
plt.title("BMI Distribution")
plt.show()
```
